### [ì‹¤ìŠµ] Scipy ë¼ì´ë¸ŒëŸ¬ë¦¬ë¡œ ì´í•­ë¶„í¬ êµ¬í˜„í•˜ê¸°

```python
#scipyíŒ¨í‚¤ì§€ì˜ specialëª¨ë“ˆì— ìˆëŠ” combí•¨ìˆ˜ ì‚¬ìš©
import numpy as np
from scipy.special import comb

def Bin(n, p):
    x_set = np.arange(n+1)

    def f(x):
        if x in x_set:
            return comb(n,x) * p**x * (1-p)**(n-x)
        else:
            return 0

    return x_set, f

n, p = 10, 5/4
X = Bin(n, p)
X
```

### [ì‹¤ìŠµ] ì´í•­ë¶„í¬ N(í‘œë³¸)ì„ ëŠ˜ë¦¬ë©´ ì •ê·œë¶„í¬í™” ëœë‹¤.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats, integrate

# ê¸°ëŒ€ê°’:í‰ê· 
def E(X, g=lambda x:x):
    x_range, f = X

    def integrand(x):
        return g(x) * f(x)

    return integrate.quad(integrand, -np.inf, np.inf)[0]

# í™•ë¥ 
def get_prob(X, a, b):
    x_range, f = X

    prob_sum = np.round(integrate.quad(f, a, b), 4)

    print(f'í™•ë¥ : {prob_sum[0]}')


def plot_prob_bi(X):
    x_set, f = X
    prob = np.array([ f(x_k) for x_k in x_set])

    fig = plt.figure(figsize=(10,6))
    ax  = fig.add_subplot(111) # (1,1,1)

    ax.bar(x_set, prob, label='prob')
    ax.set_xticks(np.append(x_set, E(X)))   #xì¶•ì˜ ê°’ ì¶œë ¥
    ax.set_ylim(0, prob.max()*1.2)          #yì¶•ì˜ í•œê³„ ì§€ì •
    ax.legend()

    plt.show()

n = 100    # nì˜ ê°œìˆ˜ë¥¼ í¬ê²Œ í•´ë³´ê¸° --> ê·¸ë˜í”„ê°€ ì¢…ëª¨ì–‘ìœ¼ë¡œ ë³€ê²½ë¨
p = 0.25   # 0.3, 0.5, 0.7 ë³€ê²½í•´ë³´ê¸°
X = Bin(n,p)   # ì´í•­ë¶„í¬ì—ì„œ  nì˜ ê°œìˆ˜ë¥¼ ëŠ˜ë¦¬ë©´?
plot_prob_bi(X)
```

### [ì‹¤ìŠµ] ì •ê·œë¶„í¬ë¥¼ ì´ìš©í•œ í™•ë¥  êµ¬í•˜ê¸°
ë‚¨ì ê³ ë“±í•™ìƒ í‚¤ì˜ í‰ê· ì´ 170cmì´ê³  í‘œì¤€í¸ì°¨ê°€ 5cmë¼ë©´, ìš°ì—°íˆ ë§Œë‚œ ë‚¨ì ê³ ë“±í•™ìƒì˜ í‚¤ëŠ” N(170, 5^2)ë¥¼ ë”°ë¥¸ ë‹¤ê³  í•  ìˆ˜ ìˆë‹¤. ì´ ì‚¬ì‹¤ì„ ë°”íƒ•ìœ¼ë¡œ ìš°ì—°íˆ ë§Œë‚œ ë‚¨ì ê³ ë“±í•™ìƒì˜ í‚¤ê°€ 165cmì´ìƒ 175cm ì´í•˜ì¼ í™•ë¥ ì€?

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats, integrate

# í™•ë¥ 
def get_prob(X, a, b):
    x_range, f = X

    prob_sum = np.round(integrate.quad(f, a, b), 4)

    print(f'í™•ë¥ : {prob_sum[0]}')

#ğ‘ ~ N(0,1) ì •ê·œí™”ë¥¼ ë”°ë¥¸ ì •ê·œë¶„í¬
def N(mu, sigma):
    x_range = [-np.inf, np.inf]
    def f(x):
        return 1 / np.sqrt(2 * np.pi * sigma**2) *\
                    np.exp(-(x-mu)**2 / (2 * sigma**2))
    return x_range, f

mu, sigma = 170, 5
X = N(mu, sigma)

a, b = 165, 175
get_prob(X, a, b)
```

```python
import numpy as np

# í‚¤ê°€ ì¤‘ë³µë˜ë©´ ì•ˆë¨, ì •ì¡°ê°€ 2ê°œì„ ì£¼ì˜!!!
data = {'íƒœì¡°':73, 'ì •ì¡°1':62, 'íƒœì¢…':45, 'ì„¸ì¢…':53, 'ë¬¸ì¢…':38,
        'ë‹¨ì¢…':16, 'ì„¸ì¡°':51, 'ì˜ˆì¢…':28, 'ì„±ì¢…':37, 'ì—°ì‚°êµ°':30,
        'ì¤‘ì¢…':56, 'ì¸ì¢…':30, 'ëª…ì¢…':33, 'ì„ ì¡°':56, 'ê´‘í•´êµ°':66,
        'ì¸ì¡°':54, 'íš¨ì¢…':40, 'í˜„ì¢…':33, 'ìˆ™ì¢…':59, 'ê²½ì¢…': 36,
        'ì˜ì¡°':82, 'ì •ì¡°2':48, 'ìˆœì¡°':44, 'í—Œì¢…':22, 'ì² ì¢…':32,
        'ê³ ì¢…':67, 'ìˆœì¢…':52}

def get_data_count(datas):
    '''ë°ì´í„°ë³„ ë¹ˆë„'''
    values = sorted( list(set(datas)) )            # ê³ ìœ ê°’ ë¦¬ìŠ¤íŠ¸
    counts = [ datas.count(i) for i in values ]   # ë¹ˆë„ìˆ˜ ë¦¬ìŠ¤íŠ¸
#     print(f'#ë°ì´í„°ë³„ ë¹ˆë„: {list(zip(values, counts))}')
    return values, counts


def get_mode_value(datas):
    '''ìµœë‹¤ ë¹ˆë„'''
    values, counts = get_data_count(datas)
    maxs = [(k, v) for k, v in zip(values, counts) if max(counts)== v]
    return maxs


names= list(data.keys())
ages = list(data.values())

x_ = np.mean(ages)               # í‰ê· ê°’
me = np.median(ages)             # ì¤‘ì•™ê°’
mo = get_mode_value(ages)        # ìµœë¹ˆê°’

print(f'#ì¡°ì„ ì˜ ì™•: {names}')
print(f'#ì™•ë“¤ì˜ ë‚˜ì´: {ages}')

values, counts = get_data_count(ages)
print(f'#ë°ì´í„°ë³„ ë¹ˆë„: {list(zip(values, counts))}')

print()
print(f'#1.ì™•ë“¤ì˜ í‰ê· ë‚˜ì´   : {round(x_)}ì„¸')

idx = ages.index( min(ages) ) # ê°’ì´ ì—¬ëŸ¬ ê°œê°€ ì•„ë‹ˆë¼ëŠ” ê°€ì •
print(f'2.ìˆ˜ëª…ì´ ê°€ì¥ ì§§ì€ ì™•: {names[idx]}, {ages[idx]}ì„¸')

idx = ages.index( max(ages) )
print(f'3.ìˆ˜ëª…ì´ ê°€ì¥ ê¸´ ì™•  : {names[idx]}, {ages[idx]}ì„¸')

print(f'4.ì™•ë“¤ì˜ ì¤‘ì•™ê°’(ë‚˜ì´): {round(me)}ì„¸')

print(f"5.ì™•ë“¤ì˜ ìµœë¹ˆê°’(ë‚˜ì´): {[f'{k}ì„¸ ' for k, v in mo ] } ")
```

### [ì‹¤ìŠµë¬¸ì œ] ì‚°ìˆ í‰ê· , ì¤‘ì•™ê°’, ìµœë¹ˆê°’ ì‚¬ì´ì˜ ê´€ê³„ ì„¤ëª…

```python
import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# ê·¸ë˜í”„ ê¸°ë³¸ ì„¤ì •
plt.rcParams['figure.figsize'] = (5, 3)        # (ê°€ë¡œ,ì„¸ë¡œ) ì¸ì¹˜ ë‹¨ìœ„
plt.rcParams['axes.unicode_minus'] = False    # ê·¸ë˜í”„ ëˆˆê¸ˆ ê°’ì—ì„œ (-)ìˆ«ìí‘œì‹œ
plt.rcParams['lines.linewidth'] = 2            # ì„  êµµê¸°
plt.rcParams['font.family'] = 'Malgun Gothic'  # í•œê¸€ í°íŠ¸ ì‚¬ìš©  (AppleGothic)


def makeFrequencyTable(datas, k=5):
    # ë„ìˆ˜ë¶„í¬í‘œ ë§Œë“¤ê¸° ì‚¬ì „ ê³¼ì •
    # k = 5                               # 1.ê³„ê¸‰ ìˆ˜
    R = round(max(datas) - min(datas), 6) # 2.R : ìµœëŒ€ì¸¡ì •ê°’ - ìµœì†Œì¸¡ì •ê°’
    w = math.ceil(R/k)                    # 3.ê³„ê¸‰ ê°„ê²©
    s = min(datas) - 0.5                  # 4.ì‹œì‘ ê³„ê¸‰ê°’
    bins = np.arange(s, max(datas)+w+1, step=w)  #ê³„ê¸‰
    index = [f'{bins[i]} ~ {bins[i+1]}' for i in range(len(bins)) if i<(len(bins)-1) ] # ê³„ê¸‰ êµ¬ê°„(index)
    hist, bins = np.histogram(datas, bins)  # ê³„ê¸‰ êµ¬ê°„ë³„ ë„ìˆ˜ ë°ì´í„°
    print(f'ê³„ê¸‰ìˆ˜(K):{k}, R:{R}, ê³„ê¸‰ê°„ê²©(w):{w}, ê³„ê¸‰ì‹œì‘ê°’(s):{s}')
    print(f'ê³„ê¸‰:{bins}')


    # ë„ìˆ˜ë¶„í¬í‘œ ë§Œë“¤ê¸°
    df = pd.DataFrame(hist, index=index, columns=['ë„ìˆ˜'])
    df.index.name = 'ê³„ê¸‰ê°„ê²©'

    df['ìƒëŒ€ë„ìˆ˜'] = [x/sum(hist) for x in hist]
    df['ëˆ„ì ë„ìˆ˜'] = [sum(hist[:i+1]) if i>0 else hist[i] for i in range(k)]
    df['ëˆ„ì ìƒëŒ€ë„ìˆ˜'] = [sum(hist[:i+1]) if i>0 else df['ìƒëŒ€ë„ìˆ˜'].values[i] for i in range(k)]
    df['ê³„ê¸‰ê°’'] = [ (bins[x]+bins[x+1])/2 for x in range(k)]

    return df


datas = [31, 27, 25, 24, 21, 44, 36, 29, 35, 28,
         16, 26, 21, 37, 26, 22, 30, 29, 40, 33,
         35, 26, 22, 42, 25, 28, 39, 30, 27, 30]

# ë„ìˆ˜ë¶„í¬í‘œ ë§Œë“¤ê¸°
df =  makeFrequencyTable(datas, k=6)
df
```

```python
def make_histogram(datas, x, y, width=10):
    fig = plt.bar(x,y, width=width, edgecolor='w') # ë§‰ëŒ€ê·¸ë˜í”„
    plt.xticks(ticks=x, labels=x)

    x_ = np.mean(datas)      # í‰ê· ê°’
    me = round(np.median(datas))    # ì¤‘ì•™ê°’
    from collections import Counter
    counts = Counter(datas)
    mo = counts.most_common(1)[0][0]  # ìµœë¹ˆê°’

    plt.axvline(x_, 0, max(y),
                c='red', label=f'mean({x_})')     # # í‰ê· ê°’ í‘œì‹œ
    plt.annotate('í‰ê· ', xy=(x_, 10), xytext=(50,10),
                 fontsize=12,
                 arrowprops=dict(facecolor='black', width=1, shrink=0.1, headwidth=5))

    plt.axvline(me, 0, max(y),
                c='green', label=f'median({me})') # ì¤‘ì•™ê°’ í‘œì‹œ

    plt.axvline(mo, 0, max(y),
                c='violet', label=f'mode({mo})')  # ìµœë¹ˆê°’ í‘œì‹œ

    plt.title('(ë„ìˆ˜)íˆìŠ¤í† ê·¸ë¨')
    plt.xlabel('ê³„ê¸‰ê°’')
    plt.ylabel('ë„ìˆ˜')

    plt.legend()
    plt.show()


# íˆìŠ¤í† ê·¸ë¨ ë§Œë“¤ê¸°
x = df['ê³„ê¸‰ê°’']
y = df['ë„ìˆ˜']
make_histogram(datas, x, y)
```

### - ë³€ë™ê³„ìˆ˜(coefficient of variation)
 - ì¸¡ì •ë‹¨ìœ„ê°€ ë™ì¼í•œ ë‘ ìë£Œì§‘ë‹¨: í‰ê· ì´ í° ì°¨ì´ë¥¼ ë³´ì´ëŠ” ë‘ ìë£Œì§‘ë‹¨
 - ì¸¡ì •ë‹¨ìœ„ê°€ ì„œë¡œ ë‹¤ë¥¸ ë‘ ìë£Œì§‘ë‹¨: ì‚°í¬ì˜ ì²™ë„ë¥¼ ë¹„êµí•  ë•Œ ë§ì´ ì‚¬ìš©í•œë‹¤.

```python
def get_cov(S, ddof=0):
    M   = np.mean(S)            # í‰ê· 
    STD = np.std(S, ddof=ddof)  # í‘œì¤€í¸ì°¨
    cov =  STD / M * 100        #ë³€ë™ê³„ìˆ˜
    print('mean: %.2f'%M, 'std: %.2f'%STD) #ì†Œìˆ˜ì  2ìë¦¬ë¡œ í‘œí˜„

    return ( np.std(S, ddof=ddof) / np.mean(S) )* 100  #ë³€ë™ê³„ìˆ˜

A = [171, 164, 167, 156, 159, 164]
B = [11.5, 12.2, 12.0, 12.4, 13.6, 10.5]

print(f'#A ë³€ë™ê³„ìˆ˜: ', round(get_cov(A, ddof=1), 2))
print(f'#B ë³€ë™ê³„ìˆ˜: ', round(get_cov(B, ddof=1), 2) )
```

### - ì™œë„(skewness)
- ë¶„í¬ì˜ ëŒ€ì¹­ì´ë‚˜ ë¹„ëŒ€ì¹­ì˜ ì •ë„ë¥¼ í‘œì‹œí•˜ëŠ” ì²™ë„

- $\alpha = 0$ :   ì™„ì „í•œ ì •ê·œë¶„í¬

### - ì²¨ë„(kurtosis)
- ë½€ì¡±í•¨ì˜ ì •ë„ë¥¼ ë‚˜íƒ€ë‚´ëŠ” ì²™ë„
- ì •ê·œë·´í¬ìƒì—ì„œ í‰ê· ì„ ì¤‘ì‹¬ìœ¼ë¡œ ì§‘ì¤‘í™”ëœ ì •ë„ë¥¼ ë‚˜íƒ€ëƒ„
- $\beta = 3$ :   ë½€ì¡±í•œ ì •ë„ê°€ ì •ê·œë¶„í¬ì™€ ê°™ë‹¤
- $\beta > 3$ :   í‘œì¤€ì •ê·œë¶„í¬ë³´ë‹¤ ì •ì ì´ ë†’ê³  ë½€ì¡±í•˜ë‹¤, ê¼¬ë¦¬ê°€ ë” ë‘êº¼ì›Œì§
- $\beta < 3$ :   í‘œì¤€ì •ê·œë¶„í¬ë³´ë‹¤ ì •ì ì´ ë‚®ê³  ì™„ë§Œí•˜ë‹¤, ê¼¬ë¦¬ê°€ ë” ì–‡ì•„ì§

```python
# ì™œë„ & ì²¨ë„ êµ¬í•˜ê¸°:  ìˆ˜í•™ì  ë°©ë²•
import numpy as np

def skew_kurtosis(data):
    mean = np.mean(data)
    var = np.var(data)
    std = np.sqrt(var)

    # skewness
    skew = np.sum((data - mean)**3) / (len(data) * std**3)
    # kurtosis
    kurt = np.sum((data - mean)**4) / (len(data) * std**4) - 3

    return skew, kurt

datas = [31, 27, 25, 24, 21, 44, 36, 29, 35, 28,
         16, 26, 21, 37, 26, 22, 30, 29, 40, 33,
         35, 26, 22, 42, 25, 28, 39, 30, 27, 30]

skew, kurt = skew_kurtosis(datas)
print("Skewness: ", skew)
print("Kurtosis: ", kurt)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def get_outlier_data(datas):
    Q1 = np.percentile(datas, 25).astype('int') # 1ì‚¬ë¶„ìœ„ (í¸ì˜ìƒ ì •ìˆ˜ë¡œ ë§Œë“¦)
    Q3 = np.percentile(datas, 75).astype('int')  # 3ì‚¬ë¶„ìœ„
    IQR = Q3 - Q1
    fl = Q1 - 1.5*IQR  # lower inner fence
    fu = Q3 + 1.5*IQR  # upper inner fence
    fL = Q1 - 3*IQR    # lower outer fence
    fU = Q3 + 3*IQR    # upper outer fence
    mid_outlier = [x for x in datas if (fL<=x<=fl) or (fu<=x<=fU)] # mid ouetlier(ë³´í†µ íŠ¹ì´ê°’)
    ext_outlier = [x for x in datas if (fL>=x) or (fU<=x)]         # extreme outlier(íŠ¹ì´ê°’)
#     print(f"ordered data     : {sorted(datas)} ")
#     print(f'lower inner fence: {fl}')
#     print(f'upper inner fence: {fu}')
#     print(f'lower outer fence: {fL}')
    print(f'ext_outlier      : {ext_outlier}')
    print(f'mid_outlier      : {mid_outlier}')



data1 = [56,  90,  88,  87,  68,  32,  76,  54, 100,  41]  # íŠ¹ì´ê°’ ë¯¸ í¬í•¨
data2 = [8, 51, 70, 86,  2, 77, 81, 75, 64, 89]    # (ë³´í†µ)íŠ¹ì´ê°’ í¬í•¨
data3 = [15, 55, 64, 95, 67, 67, 51, 58, 57, 89]   # (ë³´í†µ)íŠ¹ì´ê°’ + íŠ¹ì´ê°’ í¬í•¨
print(f'data1: {data1}')
print(f'data2: {data2}')
print(f'data3: {data3}')
print()

# outlier ê°’ ì°¾ê¸°
get_outlier_data(data1)
get_outlier_data(data2)
get_outlier_data(data3)

# ìƒìê·¸ë¦¼ ê·¸ë¦¬ê¸°
plt.boxplot([data1, data2, data3], showmeans=True)
plt.grid()
plt.show()
```

### ë‹¨ìˆœì„ í˜•íšŒê·€ëª¨í˜•(simple linear regression model) : ì§ì„ íšŒê·€ëª¨í˜•

 - $y_i = a + bx_i + e_i$   $(e_i ~ N(0,\sigma^2), i=1,2,...,n)$ <br><br>
 - $a, b$ : ëª¨íšŒê·€ê³„ìˆ˜
 - $x_i$  : ë…ë¦½ë³€ìˆ˜(ì„¤ëª…ë³€ìˆ˜), $y_i$:ì¢…ì†ë³€ìˆ˜(ë°˜ì‘ë³€ìˆ˜)
 - $e_i$  : $N(0,\sigma^2)$ì„ ë”°ë¥´ëŠ” í™•ë¥ ë³€ìˆ˜(ì˜¤ì°¨í•­) <br><br>
 -- **==> ëª¨íšŒê·€ê³„ìˆ˜ $a, b$ë¥¼ ì¶”ì •í•¨ìœ¼ë¡œì„œ íšŒê·€ì§ì„ ì„ ì¶”ì •í•  ìˆ˜ ìˆë‹¤!**

### [ì‹¤ìŠµë¬¸ì œ] ë‹¨ìˆœ íšŒê·€ë¶„ì„

```python
# 1. ë°ì´í„° í™•ì¸
import pandas as pd

cm = 2.54  # í‚¤ë°ì´í„°ë¥¼  inch --> cmë¡œ ë³€í™˜í•˜ê¸° ìœ„í•´
df = pd.read_excel('./data/Galtons Height Data_ë”¸.xlsx')
df['father'] = df.father * cm
df['daughter'] = df.daughter * cm

df
```

```python
# 2. ì‚°ì ë„(ì¶”ì •ëœ íšŒê·€ì§ì„  í‘œì‹œ) í™•ì¸
# ë°©ë²•: seaborn ì‚¬ìš©í•˜ê¸°
import matplotlib.pyplot as plt
import seaborn as sns

sns.regplot(x=df.father, y=df.daughter)
plt.xlabel('father')
plt.ylabel('daughter')
plt.show()
```

```python
# 3. ê²°ì •ê³„ìˆ˜ êµ¬í•˜ê¸°
# ë°©ë²•: statsmodels ì‚¬ìš©í•˜ê¸°
from statsmodels.formula.api import ols

model = ols('df.daughter ~ df.father', data=df).fit() #ols('ì¢…ì†ë³€ìˆ˜ëª… ~ ë…ë¦½ë³€ìˆ˜ëª…', data=ë°ì´í„°í”„ë ˆì„ê°ì²´)

print(f'ê²°ì •ê³„ìˆ˜: {model.rsquared: .4f}')

model.summary() # ëª¨ë¸ ìš”ì•½
```

```python
# ì¶”ì •ëœ íšŒê·€ì§ì„  í‘œì‹œ
# ë°©ë²•1: ì§ì ‘ ê·¸ë¦¬ê¸°
target_pred = model.predict()

plt.scatter(df.father, df.daughter, marker='o', label='$MEDV_i$ ì‹¤ì œ ê´€ì¸¡ì¹˜')
plt.plot(df.father, target_pred, color = 'g', label='$\hat{MEDV}_i$ íšŒê·€ì„ ')
plt.hlines(df.daughter.mean(), df.father.min(), df.father.max(), color = 'r', label='$MEDV_i$ í‰ê· ì„ ')
plt.legend()
plt.xlabel('df.father')
plt.ylabel('df.daughter')
plt.show()
```

```python
# ì¶”ì •ëœ íšŒê·€ì§ì„  í‘œì‹œ
# ë°©ë²•2: seaborn regplot()í•¨ìˆ˜ ì‚¬ìš©
sns.regplot(x=df.father, y=df.daughter)
plt.xlabel('ì•„ë¹ ')
plt.ylabel('ë”¸')
plt.title('ì¶”ì •íšŒê·€ì§ì„ ')
plt.show()
```

- 2.**ë…ë¦½ì„±**:   ë…ë¦½ì„±ì€ ë‹¤ì¤‘ íšŒê·€ë¶„ì„ì˜ ì¤‘ìš”í•œ ê¸°ë³¸ ê°€ì •ìœ¼ë¡œ ë…ë¦½ë³€ìˆ˜ì™€ ë‹¤ë¥¸ ë…ë¦½ë³€ìˆ˜ ê°„ì˜ **ìƒê´€ê´€ê³„ê°€ ì ì„ ê²½ìš°** ì„ í˜• íšŒê·€ ëª¨ë¸ì˜ ì˜ˆì¸¡ë ¥ì´ ì¢‹ì•„ì§„ë‹¤.

- 4.**ì •ê·œì„±**:   ì”ì°¨(residual)ë€ íšŒê·€ì§ì„ ê³¼ ê´€ì¸¡ê°’ê³¼ì˜ ì°¨ì´ì¸ë°, ì˜¤ì°¨(error)ë¼ê³ ë„ í•œë‹¤. ì´ **ì°¨ì´(ì˜¤ì°¨)ê°€ ì •ê·œì„±ì„ ë§Œì¡±**í•´ì•¼ í•œë‹¤.

### [ì‹¤ìŠµë¬¸ì œ] ì‚¬ì´í‚·ëŸ°ìœ¼ë¡œ ì„ í˜•íšŒê·€ì§ì„ (ì¶”ì •íšŒê·€ì§ì„ ) êµ¬í•˜ê¸°
ê°„ë‹¨ ë°ì´í„°ë¥¼ ì´ìš©í•˜ì—¬ ì„ í˜•íšŒê·€ì§ì„  êµ¬í•˜ê¸°

```python
import pandas as pd
import numpy as np
from sklearn import linear_model

# 1. ë°ì´í„° ì¤€ë¹„
data = {'X':[1, 1, 2, 3, 4, 4, 5, 6, 6, 8],
        'Y':[2, 3, 4, 5, 7, 8, 10, 11, 12, 15]}
df = pd.DataFrame(data)

# 2. í•™ìŠµëª¨ë¸ ì„ ì •
model = linear_model.LinearRegression()  # ì„ í˜•íšŒê·€ëª¨ë¸

# 3. í•˜ì´í¼íŒŒë¼ë¯¸í„°

# 4. í•™ìŠµ
X = [[i] for i in df.X]  # ì‚¬ì´í‚·ëŸ°ì—ì„œ ë…ë¦½ë³€ìˆ˜(X)ê°’ì€ 2ì°¨ì› í˜•íƒœë¡œ ë§Œë“¤ì–´ì¤˜ì•¼í•œë‹¤.
Y = df.Y
model.fit(X, Y)

# 5. ê²€ì¦
# y = ax + b
coef = model.coef_               # ì§ì„ ì˜ ê¸°ìš¸ê¸°(a)
intercept = model.intercept_     # ì§ì„ ì˜ ì ˆí¸(b)
score = model.score(X, Y)        # ì¶”ì •ëœ íšŒê·€ì§ì„ ì˜ ì •í™•ë„(ê²°ì •ê³„ìˆ˜ r2)
print(f"ì„ í˜•íšŒê·€ì§ì„ : y = {coef[0]:.2f}x + {intercept: .2f}")
print(f"ì„ í˜•íšŒê·€ì§ì„ ì˜ ê´€ê³„ì ìˆ˜(R2): {score: .2f}")
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import linear_model

# 1. ë°ì´í„° ì¤€ë¹„
df = pd.read_csv('./sample_data/life_expectancy.csv')
df.head(3)
```

```python
# ì»¬ëŸ¼ê°„ ìƒê´€ê´€ê³„ í™•ì¸
df.dropna(inplace=True)   # ê²°ì¸¡ì¹˜ ì œê±°

correlation_matrix = df.corr().round(2)
sns.set(rc={'figure.figsize':(22, 20)})
sns.heatmap(data=correlation_matrix, annot=True) #annot=True ìˆ«ìí‘œì‹œ
plt.show()
```

```python
# ê¸°ëŒ€ìˆ˜ëª…ê³¼ í”¼ì–´ìŠ¨ ìƒê´€ê³„ìˆ˜ ë†’ì€ ì»¬ëŸ¼ ì„ íƒ
print('ê¸°ëŒ€ìˆ˜ëª…ê³¼ ìƒê´€ê³„ìˆ˜ ë†’ì€ ë³€ìˆ˜')
c = df.corr().round(3)['Life expectancy']
c = np.abs( c ) # ìƒê´€ê³„ìˆ˜ ì ˆëŒ€ê°’

c.sort_values(ascending=True, inplace=True)  # ì˜¤ë¦„ì°¨ìˆœ ì •ë ¬
c = c[:-1]   # ìê¸° ìì‹ ì„ ì œì™¸

# ì‹œê°í™”
plt.barh(c.index, c.values)
plt.title('Correlation Coef with Life expectancy')
plt.xlabel('Features')
plt.ylabel('Correlation Coef')
plt.show()
```

```python
# ê¸°ëŒ€ìˆ˜ëª…ê³¼ í”¼ì–´ìŠ¨ ìƒê´€ê³„ìˆ˜ ë†’ì€ ì»¬ëŸ¼ ì„ íƒ
# ì¡°ê±´ ê²€ìƒ‰
c = c[c.values > 0.5]     # ìƒê´€ê³„ìˆ˜ 0.5 ì´ìƒì¸ ê°’ ì¶”ì¶œ
c.sort_values(ascending=False, inplace=True)  # ë‚´ë¦¼ì°¨ìˆœ ì •ë ¬
print(f'ê¸°ëŒ€ìˆ˜ëª…ê³¼ ìƒê´€ê´€ê³„ ë†’ì€ ì»¬ëŸ¼ : {c.index.to_list()}' )            # ì»¬ëŸ¼ëª…

# ì‹œê°í™”
sns.pairplot(df[c.index.to_list()])
plt.show()
```

```python
# ì„ í˜•íšŒê·€ëª¨ë¸ ì ìˆ˜ í™•ì¸
X = df[c.index.to_list()]
Y = df['Life expectancy']

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)

model = LinearRegression()
model.fit(X_train, y_train)
print(f'ì„ í˜•íšŒê·€ëª¨ë¸ ì ìˆ˜ = {model.score(X_test, y_test): .3f}')
```
